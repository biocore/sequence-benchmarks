```python
%matplotlib inline
import pandas as pd

comparison = pd.concat([
    pd.DataFrame.from_csv("./biopy.csv"),
    pd.DataFrame.from_csv("./cogent.csv"),
    pd.DataFrame.from_csv("./skbio023.csv"),
    pd.DataFrame.from_csv("./skbio.csv"),
], axis=1)
pd.set_option('display.float_format', lambda x: '%.2e' % x)
comparison.sort("scikit-bio master", axis=0, inplace=True)
comparison
```

```python
comparison.iloc[::-1].plot(kind='barh', logx=True, figsize=(8, 20))
```

```python
comparison.subtract(comparison['scikit-bio master'], axis=0).iloc[::-1].plot(kind='barh', figsize=(8, 20))
```
